```python
import numpy as np 
import matplotlib.pyplot as plt

x = np.zeros(100)

x[30:35]=4
x[10:22]=3
x[60:70]=3
    
x = x+np.random.randint(0,high=3,size =100)

print (len(x))
plt.plot(x)
```

```python
N = 10
G = 2
```

```python
noise_avg = []

for i in range(len(x)):
    noise_avg_lagging = sum(x[i-N-G:i])*(1/N)
    noise_avg_leading = sum(x[i+G:i+N+G])*(1/N)
    total_avg = (noise_avg_lagging + noise_avg_leading)/2
    noise_avg.append(total_avg+1)
#     print(noise_avg_lagging)
#     cut = x[i+N+G+1]

    
    
print(len(noise_avg))
plt.plot(noise_avg)
plt.plot(x)
    
```
